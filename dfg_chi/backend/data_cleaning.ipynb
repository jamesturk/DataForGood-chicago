```python
import pandas as pd
import os
import re

# Function to extract indicator name and year from filename
def extract_info_from_filename(filename):
    # Assuming the filename format is 'main_indicatorname_year.csv'
    match = re.search(r"main_(.+)_([0-9]{4}).csv", filename)
    if match:
        return match.group(1), match.group(2)
    return None, None

# Assuming all CSV files are in the 'data_directory' and named appropriately
data_directory = '/path/to/your/csv/files/'
all_dataframes = []  # List to store each year's DataFrame

# Loop through each file in the directory
for filename in os.listdir(data_directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(data_directory, filename)
        indicator_name, year = extract_info_from_filename(filename)
        if indicator_name and year:
            df = pd.read_csv(filepath)
            # Create a new DataFrame with the required columns
            new_df = pd.DataFrame({
                'census_tract_id': df['tract'].astype(int),
                'indicator_name': indicator_name,
                'value': df.iloc[:, 1].astype(int),  # Assuming value is in the second column
                'year': int(year)
            })
            all_dataframes.append(new_df)

# Concatenate all DataFrames
final_dataframe = pd.concat(all_dataframes)

# Save to a new CSV file
final_dataframe.to_csv('/path/to/save/aggregated_data.csv', index=False)
```
